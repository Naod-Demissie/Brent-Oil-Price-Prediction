## <font color = '#DF9166' size=6>**Exploratory Data Analysis**<font/><a class = 'anchor' id = 'eda'/>

```python
# Print dataset summary
eda.data_summary()
```

```python
# Plot 30-day moving average
eda.plot_moving_average(30)
```

The graph displays the 30-day moving average (red) of oil prices alongside actual prices (blue) over several decades. It shows significant price fluctuations, with notable peaks around 2008, 2012, and 2022, and sharp declines during financial crises and economic downturns.

```python
# Plot ACF and PACF
eda.plot_acf_pacf()
```

The ACF plot shows strong persistence, indicating non-stationarity, while the PACF suggests only the first few lags are significant, implying a potential AR(1) process.

```python
# Perform time series decomposition
eda.decompose_time_series()
```

The decomposition plot shows an increasing trend in oil prices, a strong seasonal pattern, and residuals with higher volatility around 2008, indicating a significant shock during that period.

```python
eda.plot_monthly_yearly_trends()  # Plot monthly and yearly trends
```

The first graph shows a seasonal price pattern, peaking mid-year, while the second graph indicates a long-term upward trend with volatility.
